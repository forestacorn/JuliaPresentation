# Plotting Graphs

## Text and terminal-based plotting
Useful for plotting from the REPL.
### UnicodePlots
```
Pkg.add("UnicodePlots")
```  
from the REPL to plot using unicode characters.
### ASCIIPlots
```
Pkg.add("ASCIIPlots")
```  
from the REPL to plot using ASCII characters.

(**Note to self:** In JuliaBox I may need to switch to kernel 0.3.12 to show this?)

````julia
using ASCIIPlots   # specify which plotting package to use

x = collect(-pi:0.2:pi)
lineplot(x, sin(x))
````

## Graphical Plotting
There are several libraries for generating figures in Julia:
- Winston
- Gadfly
- PyPlot - a Julia frontend for Matplotlib

#### Others
- Vega
- Bokeh
- And according to
https://en.wikibooks.org/wiki/Introducing_Julia/Plotting there are many more options for plotting figures too.

### PyPlot Example
```
Pkg.add("PyPlot")
```  
to install PyPlot if it is not already installed.

````julia
using PyPlot  # specify which plotting package to use.

x = collect(-5pi:0.2:5pi)
y = sin(x)./x   # note the dot.
plot(x,y, linewidth=4)
````
